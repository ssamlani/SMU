```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df=pd.read_csv(mouse_drug_data_to_load)
clinical_df=pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset
full_df = pd.merge(mouse_df, clinical_df, on="Mouse ID", how="left")
# Display the data table for preview
full_df.head
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
tumor_series = full_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].mean()
# Convert to DataFrame
tumor_df = pd.DataFrame(tumor_series)
tumor_df.reset_index(inplace=True)
# Preview DataFrame
tumor_df.head()
```

```python

# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumor_error = full_df.groupby(["Drug","Timepoint"])["Tumor Volume (mm3)"].sem()
# Convert to DataFrame
sem_df = pd.DataFrame(tumor_error)
sem_df.reset_index(inplace=True)
# Preview DataFrame
sem_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
mean_pivot = tumor_df.pivot_table(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
# Preview that Reformatting worked
mean_pivot.head()
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
ax.errorbar(mean_pivot.index, mean_pivot.Capomulin, yerr=tumor_error.Capomulin,
         fmt="o", color="r", ls="--", alpha=0.6, label="Capomulin")
ax.errorbar(mean_pivot.index, mean_pivot.Infubinol, yerr=tumor_error.Infubinol,
         fmt="^", color="b", ls="--", alpha=0.6, label="Infubinol")
ax.errorbar(mean_pivot.index, mean_pivot.Ketapril, yerr=tumor_error.Ketapril,
         fmt="s", color="g", ls="--", alpha=0.6, label="Ketapril")
ax.errorbar(mean_pivot.index, mean_pivot.Placebo, yerr=tumor_error.Placebo,
         fmt="d", color="k", ls="--", alpha=0.6, label="Placebo")
ax.set_xlim(-0.5,47)
ax.set_ylim(30, 75)
ax.set_xlabel("Timepoint")
ax.set_ylabel("Tumor Volume (mm3)")
plt.grid(axis='y')
plt.title("Tumor Response to Treatment")
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig("Tumor Response.png")
# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metast_series = full_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].mean()

metast_series 
# Convert to DataFrame
metastmean_df = pd.DataFrame(metast_series)
metastmean_df.reset_index(inplace=True)
# Preview DataFrame
metastmean_df.head()

```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metast_error = full_df.groupby(["Drug","Timepoint"])["Metastatic Sites"].sem()
# Convert to DataFrame
metastsem_df = pd.DataFrame(metast_error)
metastsem_df.reset_index(inplace=True)
# Preview DataFrame
metastsem_df.head()

```

```python
# Minor Data Munging to Re-Format the Data Frames
metasmean_pivot = metastmean_df.pivot_table(index='Timepoint',columns='Drug',values='Metastatic Sites')
# Preview that Reformatting worked
metasmean_pivot.head(20)


```

```python

# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()
ax.errorbar(metasmean_pivot.index, metasmean_pivot.Capomulin, yerr=metast_error.Capomulin,
         fmt="o", color="r", ls="--", alpha=0.6, label="Capomulin")
ax.errorbar(metasmean_pivot.index, metasmean_pivot.Infubinol, yerr=metast_error.Infubinol,
         fmt="^", color="b", ls="--", alpha=0.6, label="Infubinol")
ax.errorbar(metasmean_pivot.index, metasmean_pivot.Ketapril, yerr=metast_error.Ketapril,
         fmt="s", color="g", ls="--", alpha=0.6, label="Ketapril")
ax.errorbar(metasmean_pivot.index, metasmean_pivot.Placebo, yerr=metast_error.Placebo,
         fmt="d", color="k", ls="--", alpha=0.6, label="Placebo")


ax.set_xlabel("Treatment Duration (days)")
ax.set_ylabel("Metastatic Volume (mm3)")
plt.grid(axis='y')
plt.title("Metastatic Response to Treatment")
plt.legend(loc="best", fontsize="small", fancybox=True)
# Save the Figure
plt.savefig("Metastatic Response to Treatment.png")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
mouse_grouped_df =full_df.groupby(['Drug','Timepoint'])['Mouse ID']
count_mouse_df = mouse_grouped_df.nunique()
count_mouse_df = pd.DataFrame(count_mouse_df)
# Convert to DataFrame
count_mouse_df.reset_index(inplace=True)
count_mouse_df=count_mouse_df.rename(columns={'Mouse ID':'Mouse Count'})
# Preview DataFrame
# Preview DataFrame
count_mouse_df.head()



```

```python
# Minor Data Munging to Re-Format the Data Frames
countMousePivot =count_mouse_df.pivot_table(index='Timepoint',columns='Drug',values='Mouse Count')
# Preview the Data Frame

countMousePivot.head()
```

```python
# Generate the Plot (Accounting for percentages)

fig, ax = plt.subplots()

ax.plot(countMousePivot.index.values, (countMousePivot.Ketapril.values/countMousePivot.Ketapril.values[0] * 100), marker="s", color="g",
            alpha=0.5, label="Ketapril", ls="--")

ax.plot(countMousePivot.index.values, (countMousePivot.Capomulin.values/countMousePivot.Capomulin.values[0] * 100), marker="o", color="r",
            alpha=0.5, label="Capomulin", ls="--")

ax.plot(countMousePivot.index.values, (countMousePivot.Infubinol.values/countMousePivot.Infubinol.values[0] * 100), marker="^", color="b",
            alpha=0.5, label="Infubinol", ls="--")

ax.plot(countMousePivot.index.values, (countMousePivot.Placebo.values/countMousePivot.Placebo.values[0] * 100), marker="D", color="k",
            alpha=0.5, label="Placebo", ls="--")

ax.set_xlim(-0.5, 47)

ax.set_title("Survival During Treatment")

ax.set_xlabel("Time (days)")
ax.set_ylabel("Survival Rate (%)")

plt.legend(loc="best", fontsize="small", fancybox=True)

plt.grid(axis='both')

# Save the Figure
plt.savefig("Accounting for percentage.png")
# Show the Figure
plt.show()
```

```python
# Calculate the percent changes for each drug
percentage_change = (mean_pivot.iloc[-1]/(mean_pivot.iloc[0])-1)*100
# Display the data to confirm
percentage_change
```

```python
# Store all Relevant Percent Changes into a Tuple
pct_changes = (percentage_change["Capomulin"], 
               percentage_change["Infubinol"], 
               percentage_change["Ketapril"], 
               percentage_change["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()
ind = np.arange(len(pct_changes))  
width = 1
drugpass = ax.bar(ind[0], pct_changes[0], width, color='green')
drugfail = ax.bar(ind[1:], pct_changes[1:], width, color='red')

# Orient widths. Add labels, tick marks, etc. 
ax.set_ylabel('% Tumor Volume Change')
ax.set_title('Tumor Change Over 45 Day Treatment')
ax.set_xticks(ind + 0.5)
ax.set_xticklabels(('Capomulin', 'Infubinol', 'Ketapril', 'Placebo'))
ax.set_autoscaley_on(False)
ax.set_ylim([-30,70])
ax.grid(True)

# Use functions to label the percentages of changes
def labfail(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., 3,
                '%d%%' % int(height),
                ha='center', va='bottom', color="white")

def lablpass(rects):
    for rect in rects:
        height = rect.get_height()
        ax.text(rect.get_x() + rect.get_width()/2., -8,
                '-%d%% ' % int(height),
                ha='center', va='bottom', color="white")

# Call functions to implement the function calls
lablpass(drugpass)
labfail(drugfail)

# Save the Figure
fig.savefig("Tumor Change over 45 day Treatment.png")
# Show the Figure
fig.show()
```
